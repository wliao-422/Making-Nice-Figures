```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os as os

# os.chdir('')
fig1dat = pd.read_csv ('fig1dat.csv')
# print(fig1dat)
plt.plot(fig1dat["n"], fig1dat["NLoldrank"], 'g-o', alpha=0.5, linewidth=3)
plt.plot(fig1dat["n"], fig1dat["NLcdrank"], 'g-s', alpha=0.5, linewidth=3)
plt.xlabel('Total Number of Participants')
plt.ylabel('Cumulative Nitrogen Loss (TgN yr$^{-1}$)')

# plt.show()

# plt.plot(fig1dat["n"], fig1dat["NFoldrank"], alpha=0.5, color='tab:blue',linewidth=3)
# plt.plot(fig1dat["n"], fig1dat["NFcdrank"], alpha=0.5, color='tab:blue',linewidth=3,linestyle='-.')
# plt.xlabel('Total Number of Participants')
# plt.ylabel('Cumulative Nitrogen Input (TgN yr$^{-1}$)')

# plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5))
fig.suptitle('Effect of CM on Nitrogen Loss and Input')

ax1.plot(fig1dat["n"], fig1dat["NLoldrank"], alpha=0.5, color='tab:olive',linewidth=3)
ax1.plot(fig1dat["n"], fig1dat["NLcdrank"], alpha=0.5, color='tab:olive',linewidth=3,linestyle='--')
ax1.set(xlabel = "Total Number of Participants", ylabel = "Cumulative Nitrogen Loss (TgN yr$^{-1}$)")

ax2.plot(fig1dat["n"], fig1dat["NFoldrank"], alpha=0.5, color='tab:blue',linewidth=3)
ax2.plot(fig1dat["n"], fig1dat["NFcdrank"], alpha=0.5, color='tab:blue',linewidth=3,linestyle='--')
ax2.set(xlabel = "Total Number of Participants", ylabel = "Cumulative Nitrogen Input (TgN yr$^{-1}$)")

fig.subplots_adjust(wspace=0.2)

plt.show()
```
